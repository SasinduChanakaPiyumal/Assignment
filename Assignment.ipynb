# **Data Cleaning**

### **Outlier Detection & Remove Outliers**

**Here Glucose level cannot be less than 20 for  living person. Hence glucose level under 20 observations can be removed.**

```python
df_cleaned = df[df['Glucose'] >= 20]
```

**When Blood pressure is less than 30 is not correct for living persons. Hence we can remove rows which having blood pressure less than 30.**

```python
df_cleaned = df[df['BloodPressure'] >= 30]
```

**For BMI feasible BMI range is around 10 to around 60. Hence, BMI under 10 can be removed.**

```python
df_cleaned = df[df['BMI'] >= 10]
```

```python
df_cleaned.shape
```

### **Splitting Data**

```python
x = df_cleaned.drop(columns = 'Outcome', axis = 1)
y = df_cleaned['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

### **Scaling data**

**By studing histograms and boxplots, We can see those data has skewed distributions. Hence, we can apply minmaxscaler.**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
x_train_scaled = mm.fit_transform(x_train)
x_test_scaled = mm.transform(x_test)
```

### **Deploy Models**

```python
def model_acc(model):
    model.fit(x_train_scaled,y_train)
    acc = model.score(x_test_scaled,y_test)
    print(str(model)+'-->'+str(acc))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model_acc(lr)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model_acc(dt)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model_acc(rf)

from sklearn.svm import SVC
svm = SVC()
model_acc(svm)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
model_acc(knn)

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model_acc(nb)
```

## Accuracy Of  Models
*   LogisticRegression()-->0.7631578947368421
*   DecisionTreeClassifier()-->0.6842105263157895
*   RandomForestClassifier()-->0.7236842105263158
*   SVC()-->0.756578947368421
*   KNeighborsClassifier()-->0.7302631578947368
*   GaussianNB()-->0.7368421052631579




---



**Here, by looking at the model accuracy which has been taken by sklearn score() method (R square method), most accurate model is logistic Regression. This can be also be called as binary logistic regression. Hence, This model can be continued.**

### **Model Prediction**

### Logistic Regression

```python
y_train_pred = lr.predict(x_train_scaled)
y_test_pred = lr.predict(x_test_scaled)
```

### SVC

```python
y_train_pred_svc = svm.predict(x_train_scaled)
y_test_pred_svc = svm.predict(x_test_scaled)
```

### KNeighborsClassifier

```python
y_train_pred_kn = knn.predict(x_train_scaled)
y_test_pred_kn = knn.predict(x_test_scaled)
```

### **Model Evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Logistic Regression

```python
train_accuracy = accuracy_score(y_train,y_train_pred)
test_accuracy = accuracy_score(y_test,y_test_pred)
conf_matrix = confusion_matrix(y_test,y_test_pred)
class_report = classification_report(y_test, y_test_pred)
```

```python
print('Train_accuracy :',train_accuracy)
print('Test_accuracy :\n',test_accuracy)
print('Confusion_matrix :\n',conf_matrix)
#TT TF FT FF
print('Classification_report :\n',class_report)
```

### SVC

```python
train_accuracy = accuracy_score(y_train,y_train_pred_svc)
test_accuracy = accuracy_score(y_test,y_test_pred_svc)
conf_matrix = confusion_matrix(y_test,y_test_pred_svc)
class_report = classification_report(y_test, y_test_pred_svc)
```
